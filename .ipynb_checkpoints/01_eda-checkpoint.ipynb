```python
DATA_DIR = Path.home() / "Documents" / "Folders" / "工作" / "Kpler" / "data" / "raw"
print(DATA_DIR.resolve())
assert DATA_DIR.exists(), "Folder not found"

# Load raw (keep these untouched for audit)
pc_raw = pl.read_csv(DATA_DIR / "port_calls.csv", try_parse_dates=True)
tr_raw = pl.read_csv(DATA_DIR / "trades.csv", try_parse_dates=True)
vs_raw = pl.read_csv(DATA_DIR / "vessels.csv", try_parse_dates=True)

# Working copies (optional; clone is shallow due to COW)
df_port_calls = pc_raw.clone()
df_trades     = tr_raw.clone()
df_vessels    = vs_raw.clone()

# Parse timestamps (only if string)
def ensure_datetime(df: pl.DataFrame, cols: list[str]) -> pl.DataFrame:
    for c in cols:
        if c in df.columns and df.schema[c] == pl.Utf8:
            df = df.with_columns(pl.col(c).str.strptime(pl.Datetime, strict=False))
    return df

df_port_calls = ensure_datetime(df_port_calls, ["start_utc","end_utc"])
df_trades     = ensure_datetime(df_trades,     ["start_date_time","end_date_time"])

# Soft string normalization (trim empty/whitespace)
def normalize_strings(df: pl.DataFrame, cols: list[str]) -> pl.DataFrame:
    keep = [c for c in cols if c in df.columns]
    if not keep: return df
    df = df.with_columns([
        pl.when(pl.col(c).str.len_bytes()==0).then(None).otherwise(pl.col(c)).alias(c) for c in keep
    ])
    df = df.with_columns([pl.col(c).str.strip_chars().alias(c) for c in keep])
    return df

df_port_calls = normalize_strings(df_port_calls, ["destination"])
df_trades     = normalize_strings(df_trades,     ["product","product_family"])
df_vessels    = normalize_strings(df_vessels,    ["status","status_detail","build_country","flag_name","vessel_type"])

# Derived: stay duration (hours)
if all(c in df_port_calls.columns for c in ["start_utc","end_utc"]):
    df_port_calls = df_port_calls.with_columns(
        ((pl.col("end_utc").cast(pl.Int64) - pl.col("start_utc").cast(pl.Int64)) / 3_600_000_000).alias("dur_h")
    )

# Soft flags: is_load / is_discharge (conservative; will refine after QC)
if all(c in df_port_calls.columns for c in ["draught_change","cargo_volume","dur_h"]):
    # many zeros exist; use 10th percentile of positive cargo_volume as floor
    q10 = (
        df_port_calls.filter(pl.col("cargo_volume") > 0)
        .select(pl.col("cargo_volume").quantile(0.10))
        .item()
        if df_port_calls["cargo_volume"].max() is not None else 0
    )
    cargo_floor = float(q10) if q10 is not None else 0.0
    df_port_calls = df_port_calls.with_columns([
        ((pl.col("draught_change") >= 0.5) & (pl.col("cargo_volume") >= cargo_floor) & (pl.col("dur_h") >= 1)).alias("is_load"),
        ((pl.col("draught_change") <= -0.5) & (pl.col("cargo_volume") >= cargo_floor) & (pl.col("dur_h") >= 1)).alias("is_discharge"),
    ])

# Sort for sequencing
if all(c in df_port_calls.columns for c in ["vessel_id","start_utc"]):
    df_port_calls = df_port_calls.sort(["vessel_id","start_utc"])

# Join vessel statics (left_on vessel_id, right_on id; keep suffix to avoid collisions)
v_cols = [c for c in ["status","status_detail","build_country","build_year","flag_name","dead_weight","vessel_type"] if c in df_vessels.columns]
if v_cols:
    df_port_calls = df_port_calls.join(
        df_vessels.select(["id"] + v_cols),
        left_on="vessel_id", right_on="id", how="left", suffix="_v"
    )

# Soft de-dup: drop exact duplicates; keep last per primary key if present
before = df_port_calls.height
df_port_calls = df_port_calls.unique(maintain_order=True)
print(f"[soft] exact duplicate rows removed: {before - df_port_calls.height}")

if all(c in df_port_calls.columns for c in ["id","end_utc"]):
    df_port_calls = df_port_calls.sort("end_utc").unique(subset=["id"], keep="last")
```

## 2 - Quick Quality Check

```python
qc = {}

# missing essential keys
essential = [c for c in ("id","vessel_id","start_utc","end_utc") if c in df_port_calls.columns]
qc["rows_missing_essential"] = int(
    df_port_calls.filter(pl.any_horizontal([pl.col(c).is_null() for c in essential])).height
)

# end < start
if all(c in df_port_calls.columns for c in ("start_utc","end_utc")):
    qc["end_before_start"] = int(df_port_calls.filter(pl.col("end_utc") < pl.col("start_utc")).height)

# duplicates by quasi-key (vessel_id + start_utc)
if all(c in df_port_calls.columns for c in ("vessel_id","start_utc")):
    qc["duplicate_calls_vessel_ts"] = int(
        df_port_calls.group_by(["vessel_id","start_utc"]).len().filter(pl.col("len")>1).height
    )

# very long stays (> P99.5)
if "dur_h" in df_port_calls.columns:
    p995 = df_port_calls.select(pl.col("dur_h").quantile(0.995)).item()
    qc["very_long_stays"] = int(df_port_calls.filter(pl.col("dur_h") > p995).height)
    qc["dur_h_p99_5"] = float(p995)

print("QC summary:", qc)
```

## 3 - Data Clean and Quality Re-check

```python
# 1) 查看缺失项分布（可选）
essential = [c for c in ("id","vessel_id","start_utc","end_utc") if c in df_port_calls.columns]
missing_by_col = {
    c: int(df_port_calls.filter(pl.col(c).is_null()).height) for c in essential
}
print("Missing per essential col:", missing_by_col)

# 2) 删除关键键缺失的行（25 行）
before_rows = df_port_calls.height
df_port_calls = df_port_calls.filter(
    pl.all_horizontal([pl.col(c).is_not_null() for c in essential])
)
print(f"[surgical] dropped rows missing essential keys: {before_rows - df_port_calls.height}")

# 3) 合并 (vessel_id + start_utc) 重复（保留“更优”记录）
# 规则：优先 end_utc 晚的；若有 cargo_volume，则同等条件下优先体积更大
sort_cols = ["vessel_id", "start_utc", "end_utc"]
if "cargo_volume" in df_port_calls.columns:
    sort_cols.append("cargo_volume")

before_rows = df_port_calls.height
df_port_calls = (
    df_port_calls
    .sort(sort_cols, descending=[False]*len(sort_cols))  # 全部升序
    .unique(subset=["vessel_id","start_utc"], keep="last")  # 取组内“最后一条”
)
print(f"[surgical] merged duplicate (vessel_id,start_utc) groups, rows removed: {before_rows - df_port_calls.height}")


# 4) 重新计算 dur_h（若行数变化，保险起见重算一次）
if all(c in df_port_calls.columns for c in ("start_utc","end_utc")):
    df_port_calls = df_port_calls.with_columns(
        ((pl.col("end_utc").cast(pl.Int64) - pl.col("start_utc").cast(pl.Int64)) / 3_600_000_000).alias("dur_h")
    )

# 5) 极端停时处理：只做“标记 + 可选截尾特征”，不改原值
# compute threshold
p995 = df_port_calls.select(pl.col("dur_h").quantile(0.995)).item()

# mark + winsorized feature (do NOT modify raw dur_h)
df_port_calls = df_port_calls.with_columns([
    (pl.col("dur_h") > p995).alias("is_long_stay_outlier"),
    pl.when(pl.col("dur_h") > p995).then(p995).otherwise(pl.col("dur_h")).alias("dur_h_winsor"),
])

# quick stats
very_long_cnt = int(df_port_calls.filter(pl.col("is_long_stay_outlier")).height)
print(f"P99.5(dur_h) = {p995:.2f} h (~{p995/24:.2f} d); very_long_stays = {very_long_cnt}")
```

```python
# re-check data quality
qc2 = {}
qc2["rows_missing_essential"] = int(
    df_port_calls.filter(pl.any_horizontal([pl.col(c).is_null() for c in essential])).height
)
qc2["end_before_start"] = int(df_port_calls.filter(pl.col("end_utc") < pl.col("start_utc")).height)
qc2["duplicate_calls_vessel_ts"] = int(
    df_port_calls.group_by(["vessel_id","start_utc"]).len().filter(pl.col("len")>1).height
)
p995_new = df_port_calls.select(pl.col("dur_h").quantile(0.995)).item()
qc2["dur_h_p99_5"] = float(p995_new)
print("QC summary (after cleaning):", qc2)
```

### 3) 船型 × DWT × product_family / product（热力图）

```python
# ---- 1) Build dominant cargo per origin port_call by traded_volume ----
#    For each origin port_call, sum volume by product_family / product, then pick the largest (dominant)
if not all(c in df_trades.columns for c in ["port_call_origin_id", PF_COL, P_COL, VOL_COL]):
    raise ValueError("trades.csv must contain: port_call_origin_id, product_family, product, traded_volume")

tr_pf_vol = (
    df_trades
    .filter(pl.col(VOL_COL).is_not_null())                # drop rows with missing volume
    .group_by(["port_call_origin_id", PF_COL])
    .agg(pl.col(VOL_COL).sum().alias("vol_pf"))
)
tr_pf_dom = (
    tr_pf_vol
    .sort(["port_call_origin_id", "vol_pf"], descending=[False, True])
    .group_by("port_call_origin_id")
    .agg([pl.first(PF_COL).alias("product_family_dom"),
          pl.first("vol_pf").alias("product_family_dom_vol")])
)

tr_p_vol = (
    df_trades
    .filter(pl.col(VOL_COL).is_not_null())
    .group_by(["port_call_origin_id", P_COL])
    .agg(pl.col(VOL_COL).sum().alias("vol_p"))
)
tr_p_dom = (
    tr_p_vol
    .sort(["port_call_origin_id", "vol_p"], descending=[False, True])
    .group_by("port_call_origin_id")
    .agg([pl.first(P_COL).alias("product_dom"),
          pl.first("vol_p").alias("product_dom_vol")])
)

# ---- 2) Join dominant cargo back to port_calls (by origin = port_calls.id) ----
pc_pf = df_port_calls.join(tr_pf_dom, left_on="id", right_on="port_call_origin_id", how="left")
pc_pf = pc_pf.join(tr_p_dom, left_on="id", right_on="port_call_origin_id", how="left", suffix="_prod")

# ---- 3) DWT buckets (if DWT exists) ----
# if DWT_COL:
#     edges  = [0, 10_000, 50_000, 120_000, 200_000, 999_999_999]
#     labels = ["<10k","10–50k","50–120k","120–200k","200k+"]
#     pc_pf = pc_pf.with_columns(pl.cut(pl.col(DWT_COL), bins=edges, labels=labels).alias("dwt_bucket"))

# DWT buckets (use pl.lit for string literals to avoid ColumnNotFoundError)
if DWT_COL:
    pc_pf = pc_pf.with_columns(pl.col(DWT_COL).cast(pl.Float64))
    pc_pf = pc_pf.with_columns(
        pl.when(pl.col(DWT_COL).is_null()).then(pl.lit(None))
         .when(pl.col(DWT_COL) < 10_000).then(pl.lit("<10k"))
         .when(pl.col(DWT_COL) < 50_000).then(pl.lit("10–50k"))
         .when(pl.col(DWT_COL) < 120_000).then(pl.lit("50–120k"))
         .when(pl.col(DWT_COL) < 200_000).then(pl.lit("120–200k"))
         .otherwise(pl.lit("200k+"))
         .alias("dwt_bucket")
    )


# ---- Helpers for plotting ----
def row_share(df: pd.DataFrame) -> pd.DataFrame:
    return df.div(df.sum(axis=1).replace(0, np.nan), axis=0)

def heatmap_percent(df: pd.DataFrame, title: str, cmap="viridis"):
    ax = sns.heatmap(df, cmap=cmap, vmin=0, vmax=1)
    ax.set_title(title)
    # colorbar in %
    ax.collections[0].colorbar.formatter = mtick.PercentFormatter(xmax=1.0)
    ax.collections[0].colorbar.update_ticks()
    plt.tight_layout(); plt.show()

# ================== 4) Vessel Type × Dominant Product Family ==================
if VT_COL and "product_family_dom" in pc_pf.columns:
    # COUNT
    ctab_cnt = (
        pc_pf.filter(pl.col("product_family_dom").is_not_null())
             .group_by([VT_COL, "product_family_dom"]).len()
             .pivot(values="len", index=VT_COL, columns="product_family_dom")
             .fill_null(0)
             .to_pandas()
             .set_index(VT_COL)
    )
    plt.figure(figsize=(12,6))
    sns.heatmap(ctab_cnt, cmap="Blues")
    plt.title("Vessel Type × Dominant Product Family — COUNT")
    plt.tight_layout(); plt.show()

    # VOLUME (sum of dominant volumes)
    ctab_vol = (
        pc_pf.filter(pl.col("product_family_dom").is_not_null())
             .group_by([VT_COL, "product_family_dom"])
             .agg(pl.col("product_family_dom_vol").sum().alias("vol_sum"))
             .pivot(values="vol_sum", index=VT_COL, columns="product_family_dom")
             .fill_null(0)
             .to_pandas()
             .set_index(VT_COL)
    )
    plt.figure(figsize=(12,6))
    sns.heatmap(ctab_vol, cmap="Greens")
    plt.title("Vessel Type × Dominant Product Family — VOLUME (sum of dominant)")
    plt.tight_layout(); plt.show()

    # VOLUME SHARE (row-normalized)
    ctab_vol_share = row_share(ctab_vol)
    plt.figure(figsize=(12,6))
    heatmap_percent(ctab_vol_share, "Vessel Type × Dominant Product Family — VOLUME SHARE")

else:
    print("Skip: missing vessel_type or product_family_dom.")

# ================== 5) DWT Bucket × Dominant Product Family ==================
if "dwt_bucket" in pc_pf.columns and "product_family_dom" in pc_pf.columns:
    # COUNT
    ctab_dwt_cnt = (
        pc_pf.filter(pl.col("product_family_dom").is_not_null())
             .group_by(["dwt_bucket","product_family_dom"]).len()
             .pivot(values="len", index="dwt_bucket", columns="product_family_dom")
             .fill_null(0)
             .to_pandas()
             .set_index("dwt_bucket")
             .reindex(["<10k","10–50k","50–120k","120–200k","200k+"])
    )
    plt.figure(figsize=(12,4))
    sns.heatmap(ctab_dwt_cnt, cmap="Blues")
    plt.title("DWT Bucket × Dominant Product Family — COUNT")
    plt.tight_layout(); plt.show()

    # VOLUME
    ctab_dwt_vol = (
        pc_pf.filter(pl.col("product_family_dom").is_not_null())
             .group_by(["dwt_bucket","product_family_dom"])
             .agg(pl.col("product_family_dom_vol").sum().alias("vol_sum"))
             .pivot(values="vol_sum", index="dwt_bucket", columns="product_family_dom")
             .fill_null(0)
             .to_pandas()
             .set_index("dwt_bucket")
             .reindex(["<10k","10–50k","50–120k","120–200k","200k+"])
    )
    plt.figure(figsize=(12,4))
    sns.heatmap(ctab_dwt_vol, cmap="Greens")
    plt.title("DWT Bucket × Dominant Product Family — VOLUME")
    plt.tight_layout(); plt.show()

    # VOLUME SHARE
    ctab_dwt_vol_share = row_share(ctab_dwt_vol)
    plt.figure(figsize=(12,4))
    heatmap_percent(ctab_dwt_vol_share, "DWT Bucket × Dominant Product Family — VOLUME SHARE")
else:
    print("Skip: missing dwt_bucket or product_family_dom.")

# ================== 6) (Optional) Product-level heatmaps ==================
# 同样方式可针对 product 粒度做图；若你当前分析以 family 为主，可先跳过。
if VT_COL and "product_dom" in pc_pf.columns:
    # volume share by product
    ctab_p = (
        pc_pf.filter(pl.col("product_dom").is_not_null())
             .group_by([VT_COL, "product_dom"])
             .agg(pl.col("product_dom_vol").sum().alias("vol_sum"))
             .pivot(values="vol_sum", index=VT_COL, columns="product_dom")
             .fill_null(0)
             .to_pandas()
             .set_index(VT_COL)
    )
    # 只画占比（产品类别可能很多，绝对值图可读性弱）
    ctab_p_share = row_share(ctab_p)
    plt.figure(figsize=(12,6))
    heatmap_percent(ctab_p_share, "Vessel Type × Dominant Product — VOLUME SHARE")
```

### 5) 季节性（按月）

```python
# --- monthly aggregates ---
monthly_calls = (
    df_port_calls
    .with_columns(pl.col("start_utc").dt.truncate("1mo").alias("month"))
    .group_by("month").len()
    .sort("month")
    .to_pandas()
    .rename(columns={"len":"calls"})
)

monthly_vol = (
    df_trades
    .with_columns(pl.col("start_date_time").dt.truncate("1mo").alias("month"))
    .group_by("month")
    .agg(pl.col("traded_volume").sum().alias("traded_volume"))
    .sort("month")
    .to_pandas()
)

# --- build clean 2023 month index & merge ---
months_2023 = pd.date_range("2023-01-01", "2023-12-01", freq="MS")
monthly = (
    pd.DataFrame({"month": months_2023})
    .merge(monthly_calls, on="month", how="left")
    .merge(monthly_vol, on="month", how="left")
    .fillna(0)
)

# scale traded volume for nicer axis (e.g., in millions)
SCALE = 1e6
monthly["traded_volume_scaled"] = monthly["traded_volume"] / SCALE

# --- plot ---
fig, axes = plt.subplots(2, 1, figsize=(10, 5), sharex=True)

# 1) port calls（柱状图更直观）
axes[0].bar(monthly["month"], monthly["calls"], width=20)
axes[0].set_title("2023 Monthly Port Calls")
axes[0].set_ylabel("Calls")
axes[0].yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.0f}"))

# 2) traded volume（折线+点）
axes[1].plot(monthly["month"], monthly["traded_volume_scaled"], marker="o")
axes[1].set_title("2023 Monthly Traded Volume")
axes[1].set_ylabel("Volume (million)")  # 若 SCALE 变化，记得改这里
axes[1].yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.1f}"))

# x 轴为月份缩写
for ax in axes:
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
    ax.grid(True, axis="y", alpha=0.25)

plt.tight_layout()
plt.show()
```

#### Voyage Legs

```python
# create next port for each vessel, then random sample 100 vessels for visiualize voyage legs

# 1) 构造每艘船的相邻靠泊“线段”
pc = (
    df_port_calls
    .select(["vessel_id","start_utc","destination",
             "destination_latitude","destination_longitude"])
    .drop_nulls(["destination_latitude","destination_longitude"])
    .sort(["vessel_id","start_utc"])
    .with_columns([
        pl.col("destination_latitude").shift(-1).over("vessel_id").alias("next_lat"),
        pl.col("destination_longitude").shift(-1).over("vessel_id").alias("next_lon"),
        pl.col("destination").shift(-1).over("vessel_id").alias("next_port"),
    ])
    .filter(
        pl.col("next_lat").is_not_null() &
        pl.col("next_lon").is_not_null() &
        (
            (pl.col("destination_latitude") != pl.col("next_lat")) |
            (pl.col("destination_longitude") != pl.col("next_lon"))
        )
    )
)

# 2) 随机抽取最多 100 艘有线段的船

# eligible = pc.select("vessel_id").unique()
# n_pick = min(100, eligible.height)
# sample_ids = set(eligible.to_series().sample(n=n_pick, with_replacement=False, seed=42).to_list())

# segs = pc.filter(pl.col("vessel_id").is_in(sample_ids)).to_pandas()

# 例如 2023-10
pc_month = pc.filter(
    (pl.col("start_utc").dt.year() == 2023) 
    # &
    # (pl.col("start_utc").dt.month() == 10)
)

activity = (
    # pc_month.group_by("vessel_id").len()
    #         .sort("len", descending=True).head(100)
      pc_month.group_by("vessel_id").len()
            .sort("len", descending=True)
)
top_ids = set(activity.select("vessel_id").to_series().to_list())
segs = pc_month.filter(pl.col("vessel_id").is_in(top_ids)).to_pandas()


if segs.empty:
    raise ValueError("No segments available; try relaxing filters or increasing sample size.")

# 3) 计算视图中心
center_lat = np.mean(np.r_[segs["destination_latitude"].values, segs["next_lat"].values])
center_lon = np.mean(np.r_[segs["destination_longitude"].values, segs["next_lon"].values])

# 4) 颜色循环（每艘船一种）
palette = [
    "#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd",
    "#8c564b","#e377c2","#7f7f7f","#bcbd22","#17becf"
]
color_cycle = cycle(palette)

# 5) 画图（每艘船一条 AntPath 动画）
m = folium.Map(location=[center_lat, center_lon], zoom_start=2, tiles="CartoDB positron")

MAX_LEGS_PER_VESSEL = 60  # 为避免页面过重，可限制每艘船最多渲染多少段
for vid, dfv in segs.groupby("vessel_id"):
    dfv = dfv.sort_values("start_utc").head(MAX_LEGS_PER_VESSEL)
    # 将相邻段合并成一条连续折线：所有起点 + 最后一个终点
    coords = list(zip(dfv["destination_latitude"], dfv["destination_longitude"]))
    coords.append((dfv.iloc[-1]["next_lat"], dfv.iloc[-1]["next_lon"]))
    # folium 需要 [lat, lon] 列表
    coords = [[lat, lon] for lat, lon in coords]

    AntPath(
        locations=coords,
        color=next(color_cycle),
        weight=2,
        opacity=0.7,
        delay=800,             # 动画速度（毫秒）
        dash_array=[6, 12],    # 虚线样式
        pulse_color="#ffffff"  # 脉冲高亮色
    ).add_to(m)

out_html = "tracks_vessels_animated.html"
m.save(out_html)
print(f"Saved {out_html} with {len(sample_ids)} vessels and {len(segs)} segments.")
```
